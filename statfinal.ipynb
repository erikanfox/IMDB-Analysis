```python
import pandas as pd
import numpy as np

movies=pd.read_csv("IMDbmovies.csv")
reviews=pd.read_csv("IMDbratings.csv")
movies.sample(10)
```

```python
reviews.sample(10)
```

```python
moviesUSA=movies[movies["country"]=='USA']
moviesUSA=moviesUSA.dropna(subset=['budget'])
moviesUSA["budget"] = moviesUSA["budget"].str[2:]
moviesUSA=moviesUSA[moviesUSA["budget"].str.isdecimal()]
moviesUSA["budget"] = moviesUSA["budget"].astype(int)


```

```python
# load inflation data
inflation = pd.read_csv('CPIAUCNS.csv')
# create index multiplier
inflation['CPI_Multiplier'] = inflation['CPIAUCNS'].iloc[-1] / inflation['CPIAUCNS']
inflation["DATE"]=inflation["DATE"].str[0:4]
avg=inflation["CPI_Multiplier"].groupby(inflation["DATE"]).mean().to_frame()
inflation=avg
inflation
```

```python
moviesUSA=moviesUSA.rename(columns={"date_published": "DATE"})
moviesUSA["DATE"]=moviesUSA["DATE"].str[0:4]
moviesUSA
```

```python
df = pd.merge(moviesUSA, inflation, how='left', on='DATE')
```

```python
df["budget_adj"] = df['budget'] * df['CPI_Multiplier'] 
df["budget_adj"] = df["budget_adj"]/1000000
moviesUSA=df
```

```python
oscars=pd.read_csv("the_oscar_award.csv")
act=["ACTRESS IN A SUPPORTING ROLE","ACTRESS IN A LEADING ROLE","ACTOR","ACTRESS","ACTOR IN A SUPPORTING ROLE","ACTOR IN A LEADING ROLE"]
oscars=oscars[oscars["category"].isin(act)]
oscars.sample(10)
```

```python
moviesUSA["oscar"]=moviesUSA["actors"].apply(lambda x: any([k in x for k in oscars["name"]]))
moviesUSA["oscar"].value_counts()
    
```

```python
moviesUSA["year"]=moviesUSA["year"].astype(int)
oscars["year_film"]=oscars["year_film"].astype(int)
```

```python
drop=[]
for i in range(1931,2021):
    odata=oscars[oscars["year_film"]<=i]
    mdata=moviesUSA[moviesUSA["year"]<=i]
    moviesUSA["oscar "+str(i)]=mdata["actors"].apply(lambda x: any([k in x for k in odata["name"]]))
    drop.append("oscar "+str(i))
```
